# 雷达图案例

```python
A = [85, 70, 90, 75, 80, 88]
B = [92, 88, 75, 82, 90, 78]
C = [75, 95, 80, 85, 92, 70]
D = [80, 82, 88, 90, 75, 85]
E = [78, 85, 82, 88, 80, 90]

labels = ['学习能力','沟通能力','创新能力','实践能力','团队协作','问题解决']

plt.figure(figsize=(8,8))

n = len(labels)
angles = np.linspace(0,2*np.pi,n,endpoint=False).tolist()

A += A[:1]
B += B[:1]
C += C[:1]
D += D[:1]
E += E[:1]
angles += angles[:1]

ax = plt.subplot(111,polar=True)

ax.plot(angles,A,'o-',linewidth=2,label='A')
ax.fill(angles,A,alpha=0.25)
ax.plot(angles,B,'o-',linewidth=2,label='B')
ax.fill(angles,B,alpha=0.25)
ax.plot(angles,C,'o-',linewidth=2,label='C')
ax.fill(angles,C,alpha=0.25)
ax.plot(angles,D,'o-',linewidth=2,label='D')
ax.fill(angles,D,alpha=0.25)
ax.plot(angles,E,'o-',linewidth=2,label='E')
ax.fill(angles,E,alpha=0.25)

ax.set_theta_offset(np.pi/2)
ax.set_theta_direction(-1)

ax.set_xticks(angles[:n])
ax.set_xticklabels(labels)

plt.title('学生能力评估雷达图', fontsize=14, pad=20)
plt.legend()
plt.show()
```

# 误差棒案例

```python
cities = ['北京', '上海', '广州', '深圳']
seasons = ['春季', '夏季', '秋季', '冬季']

beijing_means = [100, 120, 90, 110] 
shanghai_means = [95, 115, 85, 105] 
guangzhou_means = [90, 110, 80, 100]  
shenzhen_means = [80, 100, 70, 90]

beijing_errors = [10, 15, 8, 12]
shanghai_errors = [8, 12, 7, 10]
guangzhou_errors = [5, 10, 6, 8]
shenzhen_errors = [6, 9, 5, 7]

x = np.arange(len(cities)) 
width = 0.2 

fig, ax = plt.subplots(figsize=(12, 7))


spring_positions = x - 1.5 * width 
ax.bar(spring_positions, 
       [beijing_means[0], shanghai_means[0], guangzhou_means[0], shenzhen_means[0]],
       width, label='春季')
ax.errorbar(spring_positions,
            [beijing_means[0], shanghai_means[0], guangzhou_means[0], shenzhen_means[0]],
            yerr=[beijing_errors[0], shanghai_errors[0], guangzhou_errors[0], shenzhen_errors[0]],
            fmt='none', ecolor='red', capsize=5)

summer_positions = x - 0.5 * width  
ax.bar(summer_positions,
       [beijing_means[1], shanghai_means[1], guangzhou_means[1], shenzhen_means[1]],
       width, label='夏季')
ax.errorbar(summer_positions,
            [beijing_means[1], shanghai_means[1], guangzhou_means[1], shenzhen_means[1]],
            yerr=[beijing_errors[1], shanghai_errors[1], guangzhou_errors[1], shenzhen_errors[1]],
            fmt='none', ecolor='red', capsize=5)

autumn_positions = x + 0.5 * width 
ax.bar(autumn_positions,
       [beijing_means[2], shanghai_means[2], guangzhou_means[2], shenzhen_means[2]],
       width, label='秋季')
ax.errorbar(autumn_positions,
            [beijing_means[2], shanghai_means[2], guangzhou_means[2], shenzhen_means[2]],
            yerr=[beijing_errors[2], shanghai_errors[2], guangzhou_errors[2], shenzhen_errors[2]],
            fmt='none', ecolor='red', capsize=5)

winter_positions = x + 1.5 * width 
ax.bar(winter_positions,
       [beijing_means[3], shanghai_means[3], guangzhou_means[3], shenzhen_means[3]],
       width, label='冬季')
ax.errorbar(winter_positions,
            [beijing_means[3], shanghai_means[3], guangzhou_means[3], shenzhen_means[3]],
            yerr=[beijing_errors[3], shanghai_errors[3], guangzhou_errors[3], shenzhen_errors[3]],
            fmt='none', ecolor='red', capsize=5)

ax.set_xlabel('城市', fontsize=12)
ax.set_ylabel('AQI值', fontsize=12)
ax.set_title('四个城市四季空气质量指数(AQI)及偏差', fontsize=15, pad=20)
ax.set_xticks(x)
ax.set_xticklabels(cities)
ax.legend()
plt.tight_layout()
plt.show()   
```

# 正切曲线案例

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

x = np.linspace(-1.47, 1.47, 1000)
y = np.tan(x)

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, y, label='正切', linewidth=2)

ax.set_xlim(-2, 2)
ax.set_ylim(-11, 11)

x_ticks = [-1/2*np.pi, -1/4*np.pi, 0, 1/4*np.pi, 1/2*np.pi]
x_labels = ['$-1/2\pi$', '$-1/4\pi$', '0', '$1/4\pi$', '$1/2\pi$']
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_labels)

y_ticks = [-10, -7.5, -5.0, -2.5, 0.0, 2.5, 5.0, 7.5, 10.0]
y_labels = ['-10.0', '-7.5', '-5.0', '-2.5', '0.0', '2.5', '5.0', '7.5', '10.0']
ax.set_yticks(y_ticks)
ax.set_yticklabels(y_labels)

ax.axhline(y=0, color='red', linestyle='--')
ax.axvline(x=0, color='red', linestyle='--')

arrow = FancyArrowPatch(
    (0.6, 0.6), (0, 0), 
    arrowstyle='-|>',
    mutation_scale=20,
    color='black',
    linewidth=2,
    zorder=10
)
ax.add_patch(arrow)
ax.text(0.6, 1.5, r'$y=\tan(x)$',color='blue',bbox=dict(facecolor='white'))
ax.text(0.6, 0.5, '奇点')

ax.grid(True, linestyle=':', alpha=0.7)
ax.set_title('正切函数曲线')
ax.set_xlabel('x轴')
ax.set_ylabel('y轴')
ax.legend()
plt.tight_layout()
plt.show()  
```

```python
# 04_temperature_precipitation_evaporation
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

month_x = np.array(['1月','2月','3月','4月','5月','6月',
                   '7月','8月','9月','10月','11月','12月'])
# 平均气温
data_tem = np.array([2.0, 2.2, 3.3, 4.5, 6.3, 10.2, 
                     20.3, 33.4, 23.0, 16.5, 12.0, 6.2])
# 降水量
data_precipitation = np.array([2.6, 5.9, 9.0, 26.4, 28.7, 70.7,  
                               175.6, 182.2, 48.7, 18.8, 6.0, 2.3])
# 蒸发量
data_evaporation = np.array([2.0, 4.9, 7.0, 23.2, 25.6, 76.7, 
                             135.6, 162.2, 32.6, 20.0, 6.4, 3.3])
fig,ax = plt.subplots()
bar_ev = ax.bar(month_x,data_evaporation,color='orange')
bar_pre = ax.bar(month_x,data_precipitation,bottom=data_evaporation,color='green')

ax.set_ylabel('水量(ml)')
ax.set_title("平均气温与降水量、蒸发量的关系")
#创建共享x轴的新绘图区域
ax_right = ax.twinx()

line = ax_right.plot(month_x,data_tem,'o-m')
ax_right.set_ylabel('气温($^\circ$C)')

plt.legend([bar_ev,bar_pre,line[0]],['蒸发量','降水量','平均气温'],shadow=True,fancybox=True)
plt.show(block=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 配置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# ========== 线型定义部分 ==========
line_styles = {
    '-': '实线 (solid)',
    '--': '虚线 (dashed)',
    '-.': '点划线 (dash-dot)',
    ':': '点线 (dotted)',
    (0, (1, 10)): '松散虚线',
    (0, (1, 1)): '密集虚线',
    (0, (5, 10)): '长间隔虚线',
    (0, (5, 5)): '中等间隔虚线',
    (0, (5, 1)): '短间隔虚线',
    (0, (3, 10, 1, 10)): '交替长短虚线',
    (0, (3, 1, 1, 1)): '点-短划线',
    (0, (3, 5, 1, 5, 1, 5)): '短-中-短虚线'
}

# ========== 标记(marker)定义部分 ==========
markers = {
    '.': '点 (point)',
    ',': '像素点 (pixel)',
    'o': '圆 (circle)',
    'v': '下三角 (triangle_down)',
    '^': '上三角 (triangle_up)',
    '<': '左三角 (triangle_left)',
    '>': '右三角 (triangle_right)',
    '1': '下三叉 (tri_down)',
    '2': '上三叉 (tri_up)',
    '3': '左三叉 (tri_left)',
    '4': '右三叉 (tri_right)',
    's': '正方形 (square)',
    'p': '五边形 (pentagon)',
    '*': '星号 (star)',
    'h': '六边形1 (hexagon1)',
    'H': '六边形2 (hexagon2)',
    '+': '加号 (plus)',
    'x': '叉号 (x)',
    'D': '菱形 (diamond)',
    'd': '窄菱形 (thin_diamond)',
    '|': '竖线 (vline)',
    '_': '横线 (hline)'
}

# ========== 打印线型和标记列表 ==========
print("=" * 50)
print("Matplotlib 线型列表")
print("=" * 50)
for k, v in line_styles.items():
    print(f"{repr(k)} : {v}")

print("\n" + "=" * 50)
print("Matplotlib 标记(marker)列表")
print("=" * 50)
for k, v in markers.items():
    print(f"{repr(k)} : {v}")

# ========== 绘制可视化图表（2列子图：左线型，右标记） ==========
fig, (ax_line, ax_marker) = plt.subplots(1, 2, figsize=(16, 8))

# ---- 左侧：绘制所有线型 ----
x_line = np.linspace(0, 10, 100)
for i, (style, desc) in enumerate(line_styles.items()):
    y = np.sin(x_line) - i * 0.5
    ax_line.plot(x_line, y, linestyle=style, linewidth=2, label=f"{repr(style)}: {desc}")
ax_line.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax_line.set_title('所有线型展示', fontsize=14)
ax_line.set_xlabel('X轴')
ax_line.set_ylabel('Y轴')

# ---- 右侧：绘制所有标记 ----
x_marker = np.linspace(0, 10, 11)  # 仅取11个点，突出标记
for i, (marker, desc) in enumerate(markers.items()):
    y = np.sin(x_marker) - i * 0.5
    ax_marker.plot(x_marker, y, marker=marker, markersize=10, linestyle='None', 
                   label=f"{repr(marker)}: {desc}")
ax_marker.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax_marker.set_title('所有标记(marker)展示', fontsize=14)
ax_marker.set_xlabel('X轴')
ax_marker.set_ylabel('Y轴')

plt.tight_layout()
plt.show()
```

# 并行多图

```python
from pyecharts.charts import Bar,Line,Grid
from pyecharts import options as opts
x_data = ['小米', '三星', '华为', '苹果', '魅族', 'vivo', 'OPPO']
y_a = [107, 36, 102, 91, 51, 113, 45]
y_b = [104, 60, 33, 138, 105, 111, 91]
# 绘制柱形图
bar = (
    Bar()
    .add_xaxis(x_data)
    .add_yaxis("商家A", y_a)
    .add_yaxis("商家B", y_b)
    .set_global_opts(title_opts=opts.TitleOpts(title="组合图表-柱形图"), 
                     yaxis_opts=opts.AxisOpts(name="销售额(万元)",name_location="center", name_gap=30))
)
# 绘制折线图
line = (
    Line()
    .add_xaxis(x_data)
    .add_yaxis("商家A", y_a)
    .add_yaxis("商家B", y_b)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="组合图表-折线图 ", pos_top="48%"),
        legend_opts=opts.LegendOpts(pos_top="48%"),
        yaxis_opts=opts.AxisOpts(name="销售额(万元)", name_location="center", name_gap=30)
    )
)
# 绘制组合图表

grid = (
    # 创建Grid类的对象,生成并行多图形式的组合图表
    Grid()
        #添加图表对象和配置项
        .add(bar,grid_opts = opts.GridOpts(pos_bottom="60%"))
        .add(line,grid_opts = opts.GridOpts(pos_top="60%"))
)
grid.render()
```

# 顺序多图

```python
from pyecharts.charts import Bar,Line,Page
from pyecharts import options as opts
x_data = ['小米', '三星', '华为', '苹果', '魅族', 'vivo', 'OPPO']
y_a = [107, 36, 102, 91, 51, 113, 45]
y_b = [104, 60, 33, 138, 105, 111, 91]
# 绘制柱形图
bar = (
    Bar()
    .add_xaxis(x_data)
    .add_yaxis("商家A", y_a)
    .add_yaxis("商家B", y_b)
    .set_global_opts(title_opts=opts.TitleOpts(title="组合图表-柱形图"), 
                     yaxis_opts=opts.AxisOpts(name="销售额(万元)",name_location="center", name_gap=30))
)
# 绘制折线图
line = (
    Line()
    .add_xaxis(x_data)
    .add_yaxis("商家A", y_a)
    .add_yaxis("商家B", y_b)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="组合图表-折线图 "),
        legend_opts=opts.LegendOpts(),
        yaxis_opts=opts.AxisOpts(name="销售额(万元)", name_location="center", name_gap=30)
    )
)
# 创建组合图表, 并在同一网页上按顺序显示柱形图和折线图
page = Page(page_title="顺序多图",interval=2)
page.add(bar,line)
page.render()
```

# 使用HourLocator设置x轴的小时刻度 and 使用DateFormatter设置x轴的日期格式

```python
# 2、股票交易日内每小时价格变化的折线图
# 给定一个股票在一个交易日内的每小时价格数据，使用Python的matplotlib库绘制一个折线图，展示该股票在交易时间内每小时的价格变化。
# # 假设的股票价格数据
# dates = ['20245289','202452810','202452811','202452812','202452813', '202452814','202452815','202452816']
# y_date = [10.5, 10.8, 10.6, 10.9, 11.1, 11.0, 10.7, 10.8]
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, HourLocator
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

dates = ['20245289','202452810','202452811','202452812','202452813', '202452814','202452815','202452816']
x_data = [datetime.strptime(d, '%Y%m%d%H') for d in dates]
y_data = [10.5, 10.8, 10.6, 10.9, 11.1, 11.0, 10.7, 10.8]
fig = plt.figure()
ax = fig.add_axes((0.1,0.1,0.8,0.8))
# 使用matplotlib绘制折线图。
ax.plot(x_data,y_data,'-o')
# 使用HourLocator设置x轴的小时刻度。
ax.xaxis.set_major_locator(HourLocator(interval=1))
# 使用DateFormatter设置x轴的日期格式。
ax.xaxis.set_major_formatter(DateFormatter('%H:00'))
# 设置图表的标题、x轴和y轴的标签。
ax.set_title('股票交易日内每小时价格变化')
ax.set_xlabel('时间')
ax.set_ylabel('价格')
plt.xticks(rotation=45)
plt.show()
```

# 棉棒图及数据标注

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体，确保中文显示正常
plt.rcParams['font.sans-serif'] = 'SimHei'
# 解决负号显示问题
plt.rcParams['axes.unicode_minus'] = False

# x轴数据（对应15个品牌的索引）
x = np.arange(1, 16)
# 燃料消耗量数据
y = np.array([5.9, 6.2, 6.7, 7.0, 7.0, 7.1, 7.2, 7.4, 
              7.5, 7.6, 7.7, 7.7, 7.7, 7.8, 7.9])
# 品牌标签
labels = np.array(['宝骏310', '宝马i3', '致享', '焕驰', '力帆530', 
                   '派力奥', '悦翔V3', '乐风RV', '奥迪A1', '威驰FS', 
                   '夏利N7', '启辰R30', '和悦A13RS', '致炫', '赛欧'])

# 创建画布
fig = plt.figure(figsize=(10, 6), dpi=80)
ax = fig.add_subplot(111)

# 绘制棉棒图
ax.stem(x, y, linefmt='--', markerfmt='o',bottom = 1,basefmt='r-.',label='棉棒图')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_ylabel('燃料消耗量(L/km)')
ax.set_ylim([0,10])
ax.legend()
ax.set_title('不同品牌轿车的燃料消耗量')
for i in range(len(x)):
    ax.text(x[i],y[i]+0.5,y[i],ha='center',fontsize=14)
plt.tight_layout()
plt.show()
```
