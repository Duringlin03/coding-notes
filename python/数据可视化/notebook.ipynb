# 线型和标记

````python
import matplotlib.pyplot as plt
import numpy as np

# 配置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# ========== 线型定义部分 ==========
line_styles = {
    '-': '实线 (solid)',
    '--': '虚线 (dashed)',
    '-.': '点划线 (dash-dot)',
    ':': '点线 (dotted)',
    (0, (1, 10)): '松散虚线',
    (0, (1, 1)): '密集虚线',
    (0, (5, 10)): '长间隔虚线',
    (0, (5, 5)): '中等间隔虚线',
    (0, (5, 1)): '短间隔虚线',
    (0, (3, 10, 1, 10)): '交替长短虚线',
    (0, (3, 1, 1, 1)): '点-短划线',
    (0, (3, 5, 1, 5, 1, 5)): '短-中-短虚线'
}

# ========== 标记(marker)定义部分 ==========
markers = {
    '.': '点 (point)',
    ',': '像素点 (pixel)',
    'o': '圆 (circle)',
    'v': '下三角 (triangle_down)',
    '^': '上三角 (triangle_up)',
    '<': '左三角 (triangle_left)',
    '>': '右三角 (triangle_right)',
    '1': '下三叉 (tri_down)',
    '2': '上三叉 (tri_up)',
    '3': '左三叉 (tri_left)',
    '4': '右三叉 (tri_right)',
    's': '正方形 (square)',
    'p': '五边形 (pentagon)',
    '*': '星号 (star)',
    'h': '六边形1 (hexagon1)',
    'H': '六边形2 (hexagon2)',
    '+': '加号 (plus)',
    'x': '叉号 (x)',
    'D': '菱形 (diamond)',
    'd': '窄菱形 (thin_diamond)',
    '|': '竖线 (vline)',
    '_': '横线 (hline)'
}

# ========== 打印线型和标记列表 ==========
print("=" * 50)
print("Matplotlib 线型列表")
print("=" * 50)
for k, v in line_styles.items():
    print(f"{repr(k)} : {v}")

print("\n" + "=" * 50)
print("Matplotlib 标记(marker)列表")
print("=" * 50)
for k, v in markers.items():
    print(f"{repr(k)} : {v}")

# ========== 绘制可视化图表（2列子图：左线型，右标记） ==========
fig, (ax_line, ax_marker) = plt.subplots(1, 2, figsize=(16, 8))

# ---- 左侧：绘制所有线型 ----
x_line = np.linspace(0, 10, 100)
for i, (style, desc) in enumerate(line_styles.items()):
    y = np.sin(x_line) - i * 0.5
    ax_line.plot(x_line, y, linestyle=style, linewidth=2, label=f"{repr(style)}: {desc}")
ax_line.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax_line.set_title('所有线型展示', fontsize=14)
ax_line.set_xlabel('X轴')
ax_line.set_ylabel('Y轴')

# ---- 右侧：绘制所有标记 ----
x_marker = np.linspace(0, 10, 11)  # 仅取11个点，突出标记
for i, (marker, desc) in enumerate(markers.items()):
    y = np.sin(x_marker) - i * 0.5
    ax_marker.plot(x_marker, y, marker=marker, markersize=10, linestyle='None', 
                   label=f"{repr(marker)}: {desc}")
ax_marker.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax_marker.set_title('所有标记(marker)展示', fontsize=14)
ax_marker.set_xlabel('X轴')
ax_marker.set_ylabel('Y轴')

plt.tight_layout()
plt.show()
````

# 误差棒案例

````python
cities = ['北京', '上海', '广州', '深圳']
seasons = ['春季', '夏季', '秋季', '冬季']

beijing_means = [100, 120, 90, 110] 
shanghai_means = [95, 115, 85, 105] 
guangzhou_means = [90, 110, 80, 100]  
shenzhen_means = [80, 100, 70, 90]

beijing_errors = [10, 15, 8, 12]
shanghai_errors = [8, 12, 7, 10]
guangzhou_errors = [5, 10, 6, 8]
shenzhen_errors = [6, 9, 5, 7]

x = np.arange(len(cities)) 
width = 0.2 

fig, ax = plt.subplots(figsize=(12, 7))


spring_positions = x - 1.5 * width 
ax.bar(spring_positions, 
       [beijing_means[0], shanghai_means[0], guangzhou_means[0], shenzhen_means[0]],
       width, label='春季')
ax.errorbar(spring_positions,
            [beijing_means[0], shanghai_means[0], guangzhou_means[0], shenzhen_means[0]],
            yerr=[beijing_errors[0], shanghai_errors[0], guangzhou_errors[0], shenzhen_errors[0]],
            fmt='none', ecolor='red', capsize=5)

summer_positions = x - 0.5 * width  
ax.bar(summer_positions,
       [beijing_means[1], shanghai_means[1], guangzhou_means[1], shenzhen_means[1]],
       width, label='夏季')
ax.errorbar(summer_positions,
            [beijing_means[1], shanghai_means[1], guangzhou_means[1], shenzhen_means[1]],
            yerr=[beijing_errors[1], shanghai_errors[1], guangzhou_errors[1], shenzhen_errors[1]],
            fmt='none', ecolor='red', capsize=5)

autumn_positions = x + 0.5 * width 
ax.bar(autumn_positions,
       [beijing_means[2], shanghai_means[2], guangzhou_means[2], shenzhen_means[2]],
       width, label='秋季')
ax.errorbar(autumn_positions,
            [beijing_means[2], shanghai_means[2], guangzhou_means[2], shenzhen_means[2]],
            yerr=[beijing_errors[2], shanghai_errors[2], guangzhou_errors[2], shenzhen_errors[2]],
            fmt='none', ecolor='red', capsize=5)

winter_positions = x + 1.5 * width 
ax.bar(winter_positions,
       [beijing_means[3], shanghai_means[3], guangzhou_means[3], shenzhen_means[3]],
       width, label='冬季')
ax.errorbar(winter_positions,
            [beijing_means[3], shanghai_means[3], guangzhou_means[3], shenzhen_means[3]],
            yerr=[beijing_errors[3], shanghai_errors[3], guangzhou_errors[3], shenzhen_errors[3]],
            fmt='none', ecolor='red', capsize=5)

ax.set_xlabel('城市', fontsize=12)
ax.set_ylabel('AQI值', fontsize=12)
ax.set_title('四个城市四季空气质量指数(AQI)及偏差', fontsize=15, pad=20)
ax.set_xticks(x)
ax.set_xticklabels(cities)
ax.legend()
plt.tight_layout()
plt.show()   
````

# 正切曲线案例

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

x = np.linspace(-1.47, 1.47, 1000)
y = np.tan(x)

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, y, label='正切', linewidth=2)

ax.set_xlim(-2, 2)
ax.set_ylim(-11, 11)

x_ticks = [-1/2*np.pi, -1/4*np.pi, 0, 1/4*np.pi, 1/2*np.pi]
x_labels = ['$-1/2\pi$', '$-1/4\pi$', '0', '$1/4\pi$', '$1/2\pi$']
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_labels)

y_ticks = [-10, -7.5, -5.0, -2.5, 0.0, 2.5, 5.0, 7.5, 10.0]
y_labels = ['-10.0', '-7.5', '-5.0', '-2.5', '0.0', '2.5', '5.0', '7.5', '10.0']
ax.set_yticks(y_ticks)
ax.set_yticklabels(y_labels)

ax.axhline(y=0, color='red', linestyle='--')
ax.axvline(x=0, color='red', linestyle='--')

arrow = FancyArrowPatch(
    (0.6, 0.6), (0, 0), 
    arrowstyle='-|>',
    mutation_scale=20,
    color='black',
    linewidth=2,
    zorder=10
)
ax.add_patch(arrow)
ax.text(0.6, 1.5, r'$y=\tan(x)$',color='blue',bbox=dict(facecolor='white'))
ax.text(0.6, 0.5, '奇点')

ax.grid(True, linestyle=':', alpha=0.7)
ax.set_title('正切函数曲线')
ax.set_xlabel('x轴')
ax.set_ylabel('y轴')
ax.legend()
plt.tight_layout()
plt.show()  
````

# 并行多图

````python
from pyecharts.charts import Bar,Line,Grid
from pyecharts import options as opts
x_data = ['小米', '三星', '华为', '苹果', '魅族', 'vivo', 'OPPO']
y_a = [107, 36, 102, 91, 51, 113, 45]
y_b = [104, 60, 33, 138, 105, 111, 91]
# 绘制柱形图
bar = (
    Bar()
    .add_xaxis(x_data)
    .add_yaxis("商家A", y_a)
    .add_yaxis("商家B", y_b)
    .set_global_opts(title_opts=opts.TitleOpts(title="组合图表-柱形图"), 
                     yaxis_opts=opts.AxisOpts(name="销售额(万元)",name_location="center", name_gap=30))
)
# 绘制折线图
line = (
    Line()
    .add_xaxis(x_data)
    .add_yaxis("商家A", y_a)
    .add_yaxis("商家B", y_b)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="组合图表-折线图 ", pos_top="48%"),
        legend_opts=opts.LegendOpts(pos_top="48%"),
        yaxis_opts=opts.AxisOpts(name="销售额(万元)", name_location="center", name_gap=30)
    )
)
# 绘制组合图表

grid = (
    # 创建Grid类的对象,生成并行多图形式的组合图表
    Grid()
        #添加图表对象和配置项
        .add(bar,grid_opts = opts.GridOpts(pos_bottom="60%"))
        .add(line,grid_opts = opts.GridOpts(pos_top="60%"))
)
grid.render()
````

````python
# 2、股票交易日内每小时价格变化的折线图
# 给定一个股票在一个交易日内的每小时价格数据，使用Python的matplotlib库绘制一个折线图，展示该股票在交易时间内每小时的价格变化。
# # 假设的股票价格数据
# dates = ['20245289','202452810','202452811','202452812','202452813', '202452814','202452815','202452816']
# y_date = [10.5, 10.8, 10.6, 10.9, 11.1, 11.0, 10.7, 10.8]
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, HourLocator
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False

dates = ['20245289','202452810','202452811','202452812','202452813', '202452814','202452815','202452816']
x_data = [datetime.strptime(d, '%Y%m%d%H') for d in dates]
y_data = [10.5, 10.8, 10.6, 10.9, 11.1, 11.0, 10.7, 10.8]
fig = plt.figure()
ax = fig.add_axes((0.1,0.1,0.8,0.8))
# 使用matplotlib绘制折线图。
ax.plot(x_data,y_data,'-o')
# 使用HourLocator设置x轴的小时刻度。
ax.xaxis.set_major_locator(HourLocator(interval=1))
# 使用DateFormatter设置x轴的日期格式。
ax.xaxis.set_major_formatter(DateFormatter('%H:00'))
# 设置图表的标题、x轴和y轴的标签。
ax.set_title('股票交易日内每小时价格变化')
ax.set_xlabel('时间')
ax.set_ylabel('价格')
plt.xticks(rotation=45)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
#从 -5 到 4.75，每隔 0.25 取一个数” → 得到一维数组
x1 = np.arange(-5, 5, 0.25)
y1 = np.arange(-5, 5, 0.25)
#使用meshgrid函数将 x1 和 y1 从一维数组转换为二维网格矩阵。转换后，x1和y1的形状相同，每个位置(i,j)分别对应网格中的 x 坐标和 y 坐标，形成平面上的网格点集合。
x1, y1 = np.meshgrid(x1, y1)
r1 = np.sqrt(x1**2 + y1**2)
#距离原点越远，z 值按正弦规律波动
z1 = np.sin(r1)
# 创建3D坐标系
fig = plt.figure()
ax_3d = fig.add_subplot(projection='3d')
# 绘制三维曲面图，指定使用冷暖色映射（蓝色到红色）表示 z 值的高低
#antialiased=False关闭抗锯齿，使曲面显示更清晰（尤其在高分辨率下）
ax_3d.plot_surface(x1,y1,z1,cmap=cm.coolwarm,linewidth=0,antialiased=False)
plt.show()
````

# 随机函数自动生成数据

在Python中，随机函数主要用于**自动生成测试数据**，是考试中绘图、数据分析、统计案例的高频考点。核心涉及两个模块：**Python内置的`random`模块**（适合生成单个/少量随机数据）和**NumPy的`np.random`模块**（适合生成批量、多维随机数组，结合Matplotlib绘图更常用）。

以下按「模块分类+核心函数+考试高频场景+实战案例」的逻辑梳理，贴合考试中“生成数据并绘图/分析”的常见题型。

---

## 一、Python内置`random`模块（基础）
`random`模块用于生成**单个或少量随机数据**，语法简单，适合基础题型（如生成随机整数、随机选择元素）。

### 核心函数（考试必记）
| 函数 | 功能 | 示例 |
|------|------|------|
| `random.random()` | 生成`[0.0, 1.0)`之间的随机浮点数 | `random.random() → 0.623456` |
| `random.randint(a, b)` | 生成`[a, b]`之间的随机整数（闭区间） | `random.randint(1, 10) → 7` |
| `random.uniform(a, b)` | 生成`[a, b]`之间的随机浮点数 | `random.uniform(0, 10) → 3.89` |
| `random.choice(seq)` | 从序列（列表/数组）中随机选一个元素 | `random.choice([1,2,3]) → 2` |
| `random.sample(seq, k)` | 从序列中随机选`k`个不重复元素 | `random.sample([1,2,3,4], 2) → [3,1]` |
| `random.shuffle(seq)` | 打乱序列的顺序（原地修改） | `lst=[1,2,3]; random.shuffle(lst) → [3,1,2]` |

### 基础示例（生成单组随机数据）
```python
import random

# 生成1个1-100的随机整数
print(random.randint(1, 100))

# 从列表中随机选一个国家（贴合你之前的绘图题）
countries = ["中国", "美国", "俄罗斯", "英国"]
print(random.choice(countries))

# 生成5个0-10的随机浮点数（手动拼接列表）
float_data = [random.uniform(0, 10) for _ in range(5)]
print(float_data)
```

## 二、NumPy的`np.random`模块（重点）
考试中**绘图、批量数据生成**的核心模块，能直接生成**多维随机数组**，完美适配Matplotlib的绘图数据要求（如散点图、柱状图、3D图的随机坐标/数值）。

### 核心函数（考试高频）
#### 1. 生成均匀分布随机数
| 函数 | 功能 | 示例 |
|------|------|------|
| `np.random.rand(d0,d1,...)` | 生成`[0.0,1.0)`的浮点数数组，参数为维度 | `np.random.rand(2,3) → 2行3列的随机浮点数` |
| `np.random.randint(low, high=None, size=None)` | 生成`[low, high)`的整数数组，`size`为维度 | `np.random.randint(0, 10, size=(3,)) → [5,2,9]` |
| `np.random.uniform(low, high, size=None)` | 生成`[low, high]`的浮点数数组 | `np.random.uniform(1, 5, size=(2,2))` |

#### 2. 生成正态分布随机数
考试中常考“模拟真实数据”（如成绩、身高、销售额），正态分布是最常用的统计分布：
| 函数 | 功能 | 示例 |
|------|------|------|
| `np.random.randn(d0,d1,...)` | 生成**标准正态分布**（均值0，方差1）的数组 | `np.random.randn(3,2)` |
| `np.random.normal(loc, scale, size)` | 生成**自定义正态分布**（`loc`=均值，`scale`=标准差） | `np.random.normal(80, 5, size=100) → 模拟100个80分左右的考试成绩` |

#### 3. 随机选择/打乱
| 函数 | 功能 | 示例 |
|------|------|------|
| `np.random.choice(seq, size=None, replace=True)` | 从序列中随机选元素，`replace=True`可重复选 | `np.random.choice([1,2,3], size=5) → [2,3,2,1,3]` |
| `np.random.shuffle(arr)` | 打乱数组的第一维（原地修改） | `arr=np.array([1,2,3]); np.random.shuffle(arr)` |

#### 4. 随机种子（`np.random.seed()`）
**考试易错点**：设置随机种子后，生成的随机数会**固定不变**，用于“复现结果”（题目要求“生成固定的随机数据”时必用）。
```python
np.random.seed(123)  # 固定种子为123
print(np.random.randint(0, 10, size=3))  # 结果固定为[2, 2, 6]
np.random.seed(123)
print(np.random.randint(0, 10, size=3))  # 再次运行仍为[2,2,6]
```

## 四、考试易错点总结
1. **`random`与`np.random`的函数差异**
   - 内置`random.randint(a, b)`是**闭区间`[a,b]`**；
   - NumPy的`np.random.randint(a, b)`是**左闭右开`[a,b)`**（考试常考参数错误，需特别注意）。

2. **多维数组的生成**
   - `np.random.rand(2,3)`直接生成2行3列数组，无需嵌套；
   - 内置`random`需通过列表推导式生成多维数据（如`[[random.random() for _ in range(3)] for _ in range(2)]`）。

3. **随机种子的作用**
   - 只有设置相同种子，随机数才会复现；未设置时，每次运行生成的随机数不同（题目要求“固定数据”时必须加`seed`）。

4. **数据维度匹配**
   绘图时，随机生成的X/Y数据维度必须一致（如散点图中`x`和`y`的长度相同），否则会报`ValueError`。
